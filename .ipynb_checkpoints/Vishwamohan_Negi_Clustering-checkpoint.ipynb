```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
transactions = pd.read_csv('Transactions.csv')
transactions
```

```python
customers = pd.read_csv('Customers.csv')
customers
```

```python
prod = pd.read_csv('Products.csv')
prod
```

```python
merged_df = pd.merge(transactions , customers , on='CustomerID' , how='inner')
merged_df
```

```python
df = pd.merge(merged_df.drop('Price',axis=1) , prod , on='ProductID' , how='inner')
df
```

```python
customer_features = df.groupby('CustomerID').agg({
        'Quantity': ['sum', 'mean'],
        'TotalValue': ['sum', 'mean'],
        'TransactionID': 'count',
        'Price': 'mean',
        'SignupDate': 'min'
    }).reset_index()
customer_features
```

```python
customer_features.columns = ['CustomerID', 'TotalQuantity', 'AvgQuantity', 
                                 'TotalSpending', 'AvgTransactionValue', 
                                 'TransactionCount', 'AvgProductPrice', 'SignupDate']
customer_features['SignupYear'] = pd.to_datetime(customer_features['SignupDate']).dt.year
customer_features['SpendingFrequency'] = customer_features['TotalSpending'] / customer_features['TransactionCount']
customer_features
```

```python
# Feature Selection
features = ['TotalQuantity', 'TotalSpending', 'TransactionCount', 
            'AvgTransactionValue', 'AvgProductPrice', 'SpendingFrequency']
X = customer_features[features]
X
```

```python
# Standardizing data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score

max_clusters=10
db_scores = []
silhouette_scores = []
    
for n_clusters in range(2, max_clusters + 1):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_scaled)
        
    db_scores.append(davies_bouldin_score(X_scaled, labels))
    silhouette_scores.append(silhouette_score(X_scaled, labels))
    
optimal_clusters = db_scores.index(min(db_scores)) + 2
db_index = min(db_scores)
silhouette = min(silhouette_scores)
print(f'Optimal no.of clusters : {optimal_clusters}')
print(f'DB Index : {db_index}')
print(f'Silhouette Score : {silhouette}')
```
