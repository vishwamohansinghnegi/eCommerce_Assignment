```python
import numpy as np
import pandas as pd
```

```python
transactions = pd.read_csv('Transactions.csv')
transactions
```

```python
customers = pd.read_csv('Customers.csv')
customers
```

```python
prod = pd.read_csv('Products.csv')
prod
```

```python
merged_df = pd.merge(transactions , customers , on='CustomerID' , how='inner')
merged_df
```

```python
df = pd.merge(merged_df.drop('Price',axis=1) , prod , on='ProductID' , how='inner')
df
```

```python
df.info()
```

```python
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])
df['SignupDate'] = pd.to_datetime(df['SignupDate'])
df.info()
```

```python
df.groupby('CustomerID').agg({
    'TotalValue': 'sum',
    'Quantity': 'sum',
    'Price': 'mean'
}).reset_index()
```

```python
# Aggregate customer-level data
customer_features = df.groupby('CustomerID').agg({
    'TotalValue': 'sum',
    'Quantity': 'sum',
    'Price': 'mean'
}).reset_index()
customer_features
```

```python
# Standardize the numerical features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
customer_features_scaled = scaler.fit_transform(customer_features[['TotalValue', 'Quantity', 'Price']])
customer_features_scaled
```

```python
# Compute similarity matrix
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(customer_features_scaled)
similarity_matrix
```

```python
# Find top 3 similar customers for first 20 customers
customer_ids = customer_features['CustomerID'].values
lookalike_results = {}

for idx, customer in enumerate(customer_ids[:20]):
    similar_customers = sorted(
        list(enumerate(similarity_matrix[idx])),
        key=lambda x: -x[1]
    )[1:4]  # Exclude self (highest similarity)

    lookalike_results[customer] = [(customer_ids[i], score) for i, score in similar_customers]

lookalike_results
```

```python
# Save results to CSV
import csv

with open("Vishwamohan_Negi_Lookalike.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerow(["CustomerID", "SimilarCustomers"])
    for key, value in lookalike_results.items():
        writer.writerow([key, value])

# Save model development script
pd.DataFrame(lookalike_results).to_csv("Vishwamohan_Negi_Lookalike.csv", index=False)
```
